```python
# Import pandas dan data
import pandas as pd
path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/Rice Production Indonesia 2020-2022.csv'
df = pd.read_csv(path)
```

```python
# Mengubah nama kolom agar mudah dimengerti
dict_rename = {'Year':'Tahun',
 'Yield.Areal(ha)':'Luas Lahan (Ha)',
 'Productivity(kw/ha)':'Produktivitas (Ton/Ha)',
 'Production.(ton)': 'Produksi (Ton)'}

df.rename(columns=dict_rename, inplace=True)
df
```

## 1. Perbandingan total produksi dari tahun 2020 sampai 2022

```python
df.groupby('Tahun')['Produksi (Ton)'].sum()
```

## 2. Perbadingan produktivitas setiap provinsi

```python
# Provinsi paling produktif
df.groupby('Provinsi')['Produktivitas (Ton/Ha)'].mean().sort_values(ascending=False).head(5)
```

```python
# Provinsi paling tidak produktif
df.groupby('Provinsi')['Produktivitas (Ton/Ha)'].mean().sort_values(ascending=False).tail(5)
```

## 3. Persentase produksi beras di Indonesia berdasarkan Provinsi di Tahun 2022

```python
# Mengambil Data Tahun 2022
df_2022 = df[df['Tahun'] == 2022].copy()
df_2022.reset_index(drop=True, inplace=True)
```

```python
# Mencari total produksi
total_produksi = df_2022['Produksi (Ton)'].sum()
```

```python
# Membuat fungsi perhitungan persentase produksi
def persentase_produksi(prod):
  return (prod / total_produksi) * 100
```

```python
# Mengaplikasi fungsi persentase produksi pada kolom
df_2022['Persentase Produksi'] = df_2022['Produksi (Ton)'].apply(persentase_produksi)
df_2022.sort_values(by='Persentase Produksi', ascending=False).head(5)
```

```python
df_2022.sort_values(by='Persentase Produksi', ascending=False).tail(5)
```
