Fungsi statistik deskriptif lainnya:

1. mean(): Menghitung rata-rata.
2. median(): Menghitung nilai tengah (median).
3. mode(): Mengembalikan nilai yang paling sering muncul.
4. std(): Menghitung standar deviasi.
5. var(): Menghitung variansi.
6. mad(): Menghitung deviasi absolut rata-rata.
7. min(): Mengembalikan nilai minimum.
8. max(): Mengembalikan nilai maksimum.
9. sum(): Menghitung jumlah total.
10. count(): Menghitung jumlah nilai non-null.
11. quantile(): Mengembalikan nilai persentil.
12. cumsum(): Menghitung jumlah kumulatif.
13. cumprod(): Menghitung hasil kali kumulatif.
14. cummin(): Menghitung nilai minimum kumulatif.
15. cummax(): Menghitung nilai maksimum kumulatif.
16. corr(): Menghitung matriks korelasi.
17. skew(): Menghitung skewness (kemiringan distribusi).
18. kurt(): Menghitung kurtosis (keruncingan distribusi).
