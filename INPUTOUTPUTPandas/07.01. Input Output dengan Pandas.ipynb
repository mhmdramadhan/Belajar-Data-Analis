# INPUT OUTPUT

Library Pandas di Python menyediakan berbagai fungsi untuk input (membaca data) dan output (menulis data) dari dan ke berbagai format file.

```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

## INPUT: Membaca Data

### 1. Membaca Data dari File CSV (Comma Separated Value)

```python
path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/provinsi_indonesia.csv'
df1 = pd.read_csv(path, sep=',')
df1
```

### 2. Membaca Data dari File Excel

```python
path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/provinsi_indonesia.xlsx'
df2 = pd.read_excel(path, sheet_name='provinsi')
df2
```

### 3. Membaca Data dari File JSON

```python
path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/provinsi_indonesia.json'
df3 = pd.read_json(path)
df3
```

## OUTPUT: Menulis Data

```python
path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/provinsi_csv_baru.csv'
df1.to_csv(path)
```

```python
path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/provinsi_excel_baru.xlsx'
df2.to_excel(path)
```

```python
path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/provinsi_json_baru.json'
df3.to_json(path)
```
