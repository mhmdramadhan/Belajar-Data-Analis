# Menggunakan fungsi pada Data/Nilai Dataframe

Menggunakan fungsi pada data di dalam DataFrame Pandas memungkinkan Anda untuk melakukan berbagai operasi transformasi dan analisis, method yang digunakan adalah apply.

```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
data = {
    'nama': ['Ansar', 'Usman', 'Hesti', 'Alex', 'Indra'],
    'gaji_dolar': [5300, 2300, 10500, 4000, 15000],
    'bonus': [100, 200, 300, 400, 500],
    'pengalaman': [1, 2, 3, 4, 5]
}
df = pd.DataFrame(data)
df
```

## Menggunakan fungsi apply

Metode apply() digunakan untuk menerapkan fungsi pada sepanjang sumbu DataFrame atau Series. Anda bisa menerapkan fungsi pada setiap baris atau kolom

```python
# Menggunakan fungsi def
def manager_or_staff(gaji):
    if gaji >= 10000:
        return 'Manager'
    else:
        return 'Staff'
```

```python
df['status'] = df['gaji_dolar'].apply(manager_or_staff)
df
```

```python
def fasilitas(df):
    if df['status'] == 'Manager':
        return 'Mobil Dinas'
    else:
        return 'Motor Dinas'
```

```python
df['fasilitas'] = df.apply(fasilitas, axis=1)
df
```

```python
def fasilitas_lain(df):
    if df['status'] == 'Manager':
        if df['pengalaman'] > 4:
            return 'Rumah Dinas'
        else:
            return 'Tidak ada'
    else:
        return 'Tidak ada'
```

```python
df['fasilitas_lain'] = df.apply(fasilitas_lain, axis=1)
df
```
