# pandas.options.display

pandas.options.display adalah salah satu atribut dalam pustaka pandas yang digunakan untuk mengatur berbagai opsi tampilan DataFrame dan objek lainnya.

### display.max_rows

Opsi ini mengatur jumlah baris maksimum yang akan ditampilkan saat sebuah DataFrame dicetak. Misalnya, jika diatur ke 10, hanya 10 baris pertama yang akan ditampilkan.

```python
pd.options.display.max_rows = 5
df_pokemon
```

### display.max_columns

Opsi ini mengatur jumlah kolom maksimum yang akan ditampilkan. Misalnya, jika diatur ke 5, hanya 5 kolom pertama yang akan ditampilkan.

```python
pd.options.display.max_columns = None
df_pokemon
```

### display.width

Opsi ini mengatur lebar maksimum dalam karakter dari tampilan DataFrame. Misalnya, jika diatur ke 80, tampilan DataFrame tidak akan melebihi 80 karakter dalam satu baris.

```python
pd.options.display.width = 1000
print(df_pokemon)
```

### display.precision

Opsi ini mengatur jumlah angka desimal yang akan ditampilkan untuk angka floating point.

```python
df_pokemon.mean(numeric_only=True)
```

```python
pd.options.display.max_rows = None
pd.options.display.precision = 4
df_pokemon.mean(numeric_only=True)
```
