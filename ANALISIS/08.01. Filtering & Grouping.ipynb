# Filtering & Grouping

Filtering dan grouping adalah dua operasi penting yang sering digunakan dalam analisis data dengan Pandas untuk memilih subset data dan menganalisis data berdasarkan kelompok tertentu.

```python
import pandas as pd

path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/Pokemon.csv'
df = pd.read_csv(path, index_col = 'Num')
df.head(6)
```

```python
df.info()
```

## Filtering

Filtering adalah proses memilih baris data yang memenuhi kondisi tertentu. Anda dapat melakukan filtering menggunakan berbagai operator perbandingan.

```python
# filter kategorikal
filter = df['Type 1'] == 'Grass' # Coba legendary
```

```python
df[filter]
```

```python
# filter numerikal
filter = df['HP'] >100
df[filter]
```

```python
# filter statistik
filter = df['Attack'] > df['Defense'] # dibalik
df[filter]
```

```python
# Dengan menggunakan operator bit-wise and dan or
filter1 = df['Attack'] >= 100
filter2 = df['Legendary']
df[filter1 & filter2]
```

```python
# Dengan menggunakan operator bit-wise and dan or
filter1 = df['Attack'] >= 100
filter2 = df['Legendary']
df[filter1 | filter2]
```

```python
# Filter dengan fungsi statistik
filter = df['Total'] == df['Total'].max()
df[filter]
```

```python
# Filter dengan fungsi statistik
filter = df['Total'] > df['Total'].mean()
df[filter]
```

## Grouping

Grouping adalah proses mengelompokkan data berdasarkan satu atau lebih kolom, dan kemudian menerapkan fungsi agregasi pada setiap kelompok

```python
# Menggunakan single column
group_type1 = df.groupby('Type 1')
```

```python
group_type1.mean(numeric_only=True)['Total']
```

```python
group_type1.count()['Total']
```

```python
# Menggunakan multiple columns
group_type12 = df.groupby(['Type 1', 'Type 2'])
group_type12.mean(numeric_only=True)['Total']
```

```python
# Menggunakan fungsi aggregat dengan lebih dari satu kolom
group_legend = df.groupby('Legendary')
group_legend.agg({'Total': 'mean', 'Attack': 'mean', 'Defense': 'mean'})
```

```python
group_type1 = df.groupby('Type 1')
group_type1.agg({'Total': 'mean', 'Attack': 'mean', 'Defense': 'mean'})
```

```python
# Menggunakan fungsi aggregat dengan lebih dari satu parameter
group_type1 = df.groupby('Type 1')
group_type1.agg({'Total': ['min', 'max']})
```

```python
# Menggunakan fungsi aggregat dengan multigroup dan multiparameter
group_type1 = df.groupby(['Type 1', 'Type 2'])
group_type1.agg({'Total': ['min', 'max']})
```

```python
# Dengan menggunakan fungsi
def range(x):
  return x.max() - x.min()

group_type1.agg({'Total': range, 'Attack': range})
```
