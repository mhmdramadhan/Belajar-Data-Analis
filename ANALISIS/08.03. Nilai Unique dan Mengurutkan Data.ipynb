# Nilai Unique dan Mengurutkan Data

## Nilai Unique

Di dalam dataframe ada beberapa method/fungsi yang bisa digunakan untuk mengetahui nilai unique dari kolom-kolomnya.

```python
import pandas as pd
pd.options.display.max_columns = None

df = pd.read_csv('/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/STEM_Salaries.csv')
df.sample(3)
```

```python
df['title'].nunique()
```

```python
df['title'].unique()
```

```python
df['title'].value_counts()
```

```python
df['Education'].value_counts()
```

## Mengurutkan Data

```python
# Mengurutkan data berdasarkan satu kolom
df.sort_values(by = 'basesalary', ascending=False) #coba ascending True
```

```python
# Mengurutkan data berdasarkan multiple columns
df.sort_values(by = ['title', 'basesalary'], ascending=True)[['company', 'title', 'basesalary']].head(20)
```

```python
# Mengurutkan berdasarkan index
df_title_index = df.set_index('title')
```

```python
df_title_index.sort_index()
```
