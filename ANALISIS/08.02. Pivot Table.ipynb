# Pivot Table

Pivot table di dalam Pandas adalah alat yang kuat untuk menganalisis dan menyusun ulang data. Pivot table memungkinkan Anda untuk merangkum data dengan cara yang mirip dengan pivot table di Excel, yakni dengan mengelompokkan data berdasarkan satu atau lebih kolom dan kemudian menghitung agregasi dari kolom lainnya.

```python
import pandas as pd

path = '/content/drive/MyDrive/Belajar Data Analisis Python/sample_data/Pokemon.csv'
df = pd.read_csv(path, index_col='Num')
df.tail(4)
```

```python
df.pivot_table(index = ['Type 1', 'Type 2'], values = ['Total', 'Attack', 'Defense'], aggfunc=['min','max'])
```

```python
df.pivot_table(index='Type 1', columns='Type 2', values='Attack', aggfunc='mean')
```
